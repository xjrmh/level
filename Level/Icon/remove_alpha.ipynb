```python
from PIL import Image
import os

path = "level_app_icon-iOS-Default-1024x1024@1x.png"
name, ext = os.path.splitext(path)
output_path = f"{name}_updated{ext}"

with Image.open(path) as img:
    print(f"Original mode: {img.mode}, size: {img.size}")
    white_bg = Image.new("RGB", img.size, (255, 255, 255))
    if img.mode == "RGBA":
        white_bg.paste(img, mask=img.split()[3])
    else:
        white_bg.paste(img)
    white_bg.save(output_path, "PNG")
    print(f"Saved as RGB (alpha removed, white fill) to {output_path}")
```
